### Exercise C.3 $\mathit{Solve\ an\ ODE\ for}$ $y(x).$ (yx_ODE.py)
$$\frac{dy}{dx} = \frac{1}{2(y-1)}, \ \ \ \ y(0)=1+\sqrt{\epsilon}, \ \ \ \ x\in[0,4]$$

For a small number, $\epsilon < 0$, use the Forward Euler method to solve the ODE with various step size in $x$: $\Delta x=1$, $\Delta x=0.25$, and $\Delta x=0.01$. Plot the numerical solution along with the analytical solution, $y(x)=1+\sqrt{x+\epsilon}$, using $1001 \ x$ coordinates. Set $\epsilon = 10^{-3}$.

For $\Delta x=1$, the approximation at $x=4$ is:

```python
p1.Forward_Euler(p1.y_prime, 0, 4, 1)[0][-1]
```

For $\Delta x=0.25$, the approximation at $x=4$ is:

```python
p1.Forward_Euler(p1.y_prime, 0, 4, 0.25)[0][-1]
```

For $\Delta x=0.01$, the approximation at $x=4$ is:

```python
p1.Forward_Euler(p1.y_prime, 0, 4, 0.01)[0][-1]
```

Graph of these solutions along with the analytical solution:

```python
p1.plot()
```

A symbolic solution can be found using sympy instead of numpy:

```python
p1.sympy_solution()
```

Using the initial condition, $y(0)=1+\sqrt{\epsilon}$, the constant, C1, can be derived:
$$f(0)=y(0)=1+\sqrt{\epsilon}=\sqrt{C1+0}+1$$
$$\Rightarrow \epsilon = C1$$
$$\Rightarrow y(x)=1+\sqrt{\epsilon + x}$$

This particular ODE is difficult to solve because with a step size of $\Delta x=1$, the solution is very innacurate, even when we use $1001 \ x$ coordinates. To solve this numerically by hand, we would need a small step size and a very large number of iterations.
